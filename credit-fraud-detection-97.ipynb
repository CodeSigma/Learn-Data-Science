```python
tp = 0
fp = 0
tn = 0
fn = 0
y_test = list(y_test)
res = rf.predict(X_test)

for i in range(len(res)):
    if y_test[i] == 1:
        if res[i] == 1:
            tp+=1
        else:
            fn+=1
    else:
        if res[i] == 0:
            tn+=1
        else:
            fp+=1
```
