```python
#lambda function with multiple columns
#axis = 1 means column, and axis = 0 means row

df['new_and_cheap'] = df.apply(lambda x: 'yes' if x['price'] <10000 and x['age'] < 10 else 'no', axis=1)
df.head().new_and_cheap
```

Sometimes a lot of null on a column may hurt the models, so we have to remove the column that has a amount of nulls that higher than threshold

```python
threshold = len(data)*.5
data.dropna(thresh=threshold, axis=1).shape
```

We can also drop the rows when the nulls is higher than a certain treshold

```python
data.dropna(thresh=21, axis=0)
```

Now, let's clean the cylinders column

```python
data.cylinders.head(10)
```

```python
#there are nulls, that's why the dtype is object
data.cylinders.isnull().any()
```

```python
data.cylinders.value_counts()
```

```python
#remove the string so it makes easier to put the column into models
#strip() is for removing spaces
data.cylinders = data.cylinders.apply(lambda x: str(x).lower().replace("cylinders",'').strip())
data.cylinders
```

```python
#convert into numeric
data.cylinders = pd.to_numeric(data.cylinders, errors = 'coerce')
```

```python
#replace the null into mean
data.cylinders.fillna(data.cylinders.mean(), inplace=True)
```

```python
data.cylinders
```

```python
data.cylinders.hist(bins = 15)
```

it seems the distribution is not right, there's some super high numbers that causing that hurt the histogram
let's clean those noise!

```python
#removing outliers, so we can get a better distribution
outliers = data[(data.odometer < data.odometer.quantile(.995)) & (data.odometer > data.odometer.quantile(.005))]
outliers
```

```python
outliers.odometer.hist()
```
